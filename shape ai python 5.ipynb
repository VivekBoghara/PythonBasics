```python
for i in range(10):
    if i==8:
        break   #This will terminate the loop from i=8 so that's why i=8 and other after that will not be printed. only 0 to 7 will be come
    print(i)    
```

```python
for i in range(10):
    if i==8:
        continue    #this continue will take out i=8 from the loop but other all will be continue. So all 0 to 9 will be printed except 8 number
    print(i)    
```

```python
** Convert the strings all uppercase into lowercase and lowercase into uppercase
Www.HackErRAnk.Com -> wWW.hACKeTraNK.cOM
```

```python
var1="Www.HackErRAnk.Com"
for i in var1:
    if i.isupper():
        print(i.lower(),end="")
    else:
        print(i.upper(),end="")
        
print("\n") 
print(var1.swapcase())
```

**A string is called as palindrome string if it's reverse will be as same as before. find that the following string is a palindrome or not.

```python
var1=input()
var2=reversed(var1)  #This reversed() is an inbuilt method in Python programming language that reverses objects of list in place. Returns: The reverse() method does not return any value but reverse the given object from the list.
# This reverse returns in form of a list thats why you have to compare both var1 and var2 in form of list, otherwise it can't be camparable
# print(type(var2))
if list(var1) == list(var2):   
    print("This is a palindrome")
else:
    print("not a palindrome")
```
